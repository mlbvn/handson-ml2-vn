## Chương 4: Huấn luyện Mô hình
Phương trình 4-1: Dự đoán của mô hình Hồi quy Tuyến tính

$\hat{y}=\theta_{0}+\theta_{1} x_{1}+\theta_{2} x_{2}+\cdots+\theta_{n} I_{n}$

---

Phương trình 4-2: Dự đoán của mô hình Hồi quy Tuyến tính (dạng vector)

$\hat{y}={h}_{\boldsymbol{\theta}}\left(\mathbf{x}\right)=\boldsymbol{\theta}\cdotp \mathbf{x}$

---
*Trích ghi chú nằm ở phần Hồi quy Tuyến tính, trang số 100 (Tập 1):*

Trong Học Máy, vector thường được biểu diễn dưới dạng *vector cột*, là mảng 2D với một cột duy nhất. 
Nếu $\boldsymbol{\theta}$ và $\mathbf{x}$ là các vector cột, thì giá trị dự đoán được tính là $\hat{y}={\boldsymbol{\theta}}^{\top }\mathbf{x}$, 
trong đó $ {\boldsymbol{\theta}}^{\top}$ là chuyển vị của $\boldsymbol{\theta}$ (một ma trận hàng, thay vì ma trận cột) và 
${\boldsymbol{\theta}}^{\top }\mathbf{x}$ là phép nhân ma trận của ${\boldsymbol{\theta}}^{\top}$ và $\mathbf{x}$.
Tất nhiên nó vẫn trả về cùng kết quả, chỉ khác là được biểu diễn dưới dạng ma trận một phần tử (*single cell matrix*) 
thay vì số vô hướng. Trong cuốn sách này, tác giả sử dụng ký hiệu trên để tránh việc chuyển đổi giữa tích vô hướng và phép 
nhân ma trận.

---

Phương trình 4-3: Hàm chi phí MSE cho mô hình Hồi quy Tuyến tính

$\text{MSE}(\mathbf{X},{h}_{\boldsymbol{\theta}})={\displaystyle \frac{1}{m}}\sum _{i=1}^{m}{({\boldsymbol{\theta}}^{\top }{\mathbf{x}}^{\left(i\right)}-{y}^{\left(i\right)})}^{2}$

---

Phương trình 4-4: Phương trình Pháp tuyến

$\hat{\boldsymbol{\theta}}={\left({\mathbf{X}}^{\top }\mathbf{X}\right)}^{-1} {\mathbf{X}}^{\top } \mathbf{y}$

---
*Trích đoạn văn nằm ở phần Phương trình Pháp tuyến, trang số 103 (Tập 1):*

Hàm này tính $\hat{\boldsymbol{\theta}}={\mathbf{X}}^{+}\mathbf{y}$, trong đó $\mathbf{X}^{+}$ là *ma trận giả nghịch đảo (pseudoinverse)* 
của $\mathbf{X}$ (cụ thể là nghịch đảo Moore-Penrose).

---
*Trích đoạn văn nằm ở phần Phương trình Pháp tuyến, trang số 103 (Tập 1):*

Bản thân ma trận nghịch đảo tổng quát này có thể được tính bằng một kỹ thuật phân rã ma trận tiêu chuẩn là *Phân tích Giá trị Suy biến* 
(*Singular Value Decomposition* -- SVD). Kỹ thuật này có thể phân rã ma trận chứa tập huấn luyện $\mathbf{X}$ thành tích của ba ma trận 
$\mathbf{U} \boldsymbol{\Sigma} \mathbf{V}^{\top}$ (tham khảo $\texttt{numpy.linalg.svd()}$). Ma trận giả nghịch đảo được tính bằng công 
thức ${\mathbf{X}}^{+}=\mathbf{V}{\boldsymbol{\Sigma }}^{+}{\mathbf{U}}^{\top}$. Để tính ma trận ${\boldsymbol{\Sigma}}^{+}$, thuật toán 
lấy $\boldsymbol{\Sigma}$ và thay giá trị 0 vào tất cả các giá trị nhỏ hơn một ngưỡng (rất nhỏ) nhất định, sau đó thay thế tất cả các giá 
trị khác 0 bằng nghịch đảo của chúng và cuối cùng chuyển vị ma trận thu được. Hướng tiếp cận này hiệu quả hơn so với việc giải Phương trình 
Pháp tuyến, đồng thời xử lý tốt hơn các trường hợp đặc biệt. Thật vậy, ta không giải được Phương trình Pháp tuyến nếu 
$\mathbf{X}^{\top} \mathbf{X}$ không có nghịch đảo (tức là ma trận suy biến -- singular matrix), chẳng hạn như khi $m < n$ hoặc khi một số 
đặc trưng là dư thừa, trong khi ma trận giả nghịch đảo thì luôn xác định được.

---
Phương trình 4-5: Đạo hàm riêng của hàm chi phí

${\displaystyle \frac{\partial }{\partial {\theta }_{j}}}\text{MSE}\left(\boldsymbol{\theta}\right)={\displaystyle \frac{2}{m}}\sum _{i=1}^{m}({\boldsymbol{\theta}}^{\top }{\mathbf{x}}^{\left(i\right)}-{y}^{\left(i\right)})\phantom{\rule{0.166667em}{0ex}}{x}_{j}^{\left(i\right)}$

---
Phương trình 4-6: Vector gradient của hàm chi phí

$\def\arraystretch{1.7}{\nabla }_{\boldsymbol{\theta}}\phantom{\rule{0.166667em}{0ex}}\text{MSE}\left(\boldsymbol{\theta}\right)=\left(\begin{array}{c}\frac{\partial }{\partial {\theta }_{0}}\text{MSE}\left(\boldsymbol{\theta}\right)\\ \frac{\partial }{\partial {\theta }_{1}}\text{MSE}\left(\boldsymbol{\theta}\right)\\ \vdots \\ \frac{\partial }{\partial {\theta }_{n}}\text{MSE}\left(\boldsymbol{\theta}\right)\end{array}\right)={\displaystyle \frac{2}{m}}{\mathbf{X}}^{\top }(\mathbf{X}\boldsymbol{\theta}-\mathbf{y})$

---
Phương trình 4-7: Bước cập nhật của Hạ Gradient

${\boldsymbol{\theta}}^{\left(\text{bước tiếp theo}\right)}=\boldsymbol{\theta}-\eta {\nabla }_{\boldsymbol{\theta}}\hspace{1pt}\text{MSE}\left(\boldsymbol{\theta}\right)$

---
*Trích đoạn văn nằm ở phần Hồi quy Ridge, trang số 119 (Tập 1):*

*Hồi Quy Ridge* (*Ridge Regression*, còn gọi là *điều chuẩn Tikhonov*) là một phiên bản được điều chuẩn của Hồi quy Tuyến tính: *tham số điều chuẩn* (*regularization term*) $\alpha \sum_{i=1}^{n} \theta_{i}^{2}$ được thêm vào hàm chi phí.

---
Phương trình 4-8: Hàm chi phí của Hồi quy Ridge

$J\left(\boldsymbol{\theta}\right)=\text{MSE}\left(\boldsymbol{\theta}\right)+\alpha \frac{1}{2}\sum _{i=1}^{n}{{\theta }_{i}}^{2} $

---
Phương trình 4-9: Phương trình dạng đóng của Hồi quy Ridge

$\hat{\boldsymbol{\theta}}={({\mathbf{X}}^{\top }\mathbf{X}+\alpha \mathbf{A})}^{-1} {\mathbf{X}}^{\top } \mathbf{y}$

---
Phương trình 4-10: Hàm chi phí của Hồi quy Lasso

$J\left(\boldsymbol{\theta}\right)=\text{MSE}\left(\boldsymbol{\theta}\right)+\alpha \sum _{i=1}^{n}\left|{\theta }_{i}\right| $

---
Phương trình 4-11: Vector subgradient của Hồi quy Lasso

$g(\boldsymbol{\theta},J)={\nabla }_{\boldsymbol{\theta}}\hspace{1pt}\text{MSE}\left(\boldsymbol{\theta}\right)+\alpha \left(\begin{array}{c}sign\left({\theta }_{1}\right)\\ sign\left({\theta }_{2}\right)\\ \vdots \\ sign\left({\theta }_{n}\right)\end{array}\right)  \text{với }sign\left({\theta }_{i}\right)=\left\{\begin{array}{cc}-1& \text{nếu }{\theta }_{i}<0\\ 0& \text{nếu }{\theta }_{i}=0\\ +1& \text{nếu }{\theta }_{i}>0\end{array}\right. $

---
Phương trình 4-12: Hàm chi phí của Elastic

$J\left(\boldsymbol{\theta}\right)=\text{MSE}\left(\boldsymbol{\theta}\right)+r\alpha \sum _{i=1}^{n}\left|{\theta }_{i}\right|+\frac{1-r}{2}\alpha \sum _{i=1}^{n}{{\theta }_{i}}^{2}$

---
Phương trình 4-13: Xác suất ước lượng của mô hình Hồi quy Logistic (dạng vector)

$\hat{p}={h}_{\boldsymbol{\theta}}\left(\mathbf{x}\right)=\sigma \left({\mathbf{x}}^{\top }\boldsymbol{\theta}\right)$

---
Phương trình 4-14: Hàm logistic

$\sigma \left(t\right)={\displaystyle \frac{1}{1+\exp(-t)}}$

---
*Trích đoạn văn nằm ở phần "Ước lượng Xác suất", trang 127 (Tập 1):*

Khi mô hình Hồi quy Logistic đã ước lượng xác suất mẫu $\mathbf{x}$ thuộc lớp dương $\hat{p}=h_{\boldsymbol{\theta}}(\mathbf{x})$, nó có thể đưa ra dự đoán một cách dễ dàng (tham khảo Phương trình 4-15).

Phương trình 4-15: Dự đoán của mô hình Hồi quy Logistic

$ \hat{y}=\left\{\begin{array}{cc}0 & \text{nếu}\phantom{\rule{4.pt}{0ex}}\hat{p}<0.5 \\1 & \text{nếu}\phantom{\rule{4.pt}{0ex}}\hat{p}\ge 0.5 \end{array}\right. $

---
Phương trình 4-16: Hàm chi phí của một mẫu huấn luyện

$c\left(\boldsymbol{\theta}\right)=\left\{\begin{array}{ll}-\mathrm{log}\left(\hat{p}\right)& \text{nếu }y=1\\ -\mathrm{log}(1-\hat{p})& \text{nếu }y=0\end{array}\right. $

---
Phương trình 4-17: Hàm chi phí Hồi quy Logistic

$J\left(\boldsymbol{\theta}\right)=-\frac{1}{m}\sum _{i=1}^{m}\left[{y}^{\left(i\right)}\log\left({\hat{p}}^{\left(i\right)}\right)+(1-{y}^{\left(i\right)})\log\left(1-{\hat{p}}^{\left(i\right)}\right)\right] $

---
Phương trình 4-18: Đạo hàm riêng của hàm chi phí logistic

${\displaystyle \frac{\partial }{\partial {\theta }_{j}}}\text{J}\left(\boldsymbol{\theta}\right)={\displaystyle \frac{1}{m}}\sum _{i=1}^{m}\left(\sigma \left({\boldsymbol{\theta}}^{\top }{\mathbf{x}}^{\left(i\right)}\right)-{y}^{\left(i\right)}\right)\phantom{\rule{0.166667em}{0ex}}{x}_{j}^{\left(i\right)} $

---
Phương trình 4-19: Điểm số Softmax cho lớp $k$

${s}_{k}\left(\mathbf{x}\right)={\mathbf{x}}^{\top }{\boldsymbol{\theta}}^{\left(k\right)} $

---
Phương trình 4-20: Hàm softmax

${\hat{p}}_{k}=\sigma {\left(\mathbf{s}\left(\mathbf{x}\right)\right)}_{k}={\displaystyle \frac{\exp\left({s}_{k}\left(\mathbf{x}\right)\right)}{\sum _{j=1}^{K}\exp\left({s}_{j}\left(\mathbf{x}\right)\right)}} $

---
Phương trình 4-21: Dự đoán của bộ phân loại Hồi quy Softmax

$\hat{y}=\underset{k}{\text{argmax}}\phantom{\rule{0.166667em}{0ex}}\sigma {\left(\mathbf{s}\left(\mathbf{x}\right)\right)}_{k}=\underset{k}{\text{argmax}}\phantom{\rule{0.166667em}{0ex}}{s}_{k}\left(\mathbf{x}\right)=\underset{k}{\text{argmax}}\phantom{\rule{0.166667em}{0ex}}\left({\left({\boldsymbol{\theta}}^{\left(k\right)}\right)}^{\top }\mathbf{x}\right) $

---
Phương trình 4-22: Hàm chi phí entropy chéo

$J\left(\boldsymbol{\theta}\right)=-\frac{1}{m}\sum _{i=1}^{m}\sum _{k=1}^{K}{y}_{k}^{\left(i\right)}\mathrm{log}\left({\hat{p}}_{k}^{\left(i\right)}\right)$

---
Phương trình 4-23: Vector gradient của entropy chéo cho lớp k

${\nabla }_{{\boldsymbol{\theta}}^{\left(k\right)}}\phantom{\rule{0.166667em}{0ex}}J\left(\boldsymbol{\theta}\right)={\displaystyle \frac{1}{m}}\sum _{i=1}^{m}\left({\hat{p}}_{k}^{\left(i\right)}-{y}_{k}^{\left(i\right)}\right){\mathbf{x}}^{\left(i\right)} $

## Chương 5: Máy Vector Hỗ trợ

Phương trình 5-1: Hàm tương tự Gauss

${\displaystyle {\phi }_{\gamma }(\mathbf{x},\ell )}={\displaystyle \exp\left({\displaystyle -\gamma {\left\Vert \mathbf{x}-\ell \right\Vert}^{2}}\right)}$

---
Phương trình 5-2: Dự đoán của bộ phân loại SVM tuyến tính

$\hat{y}=\left\{\begin{array}{ll}0 & \text{nếu}\phantom{\rule{4.pt}{0ex}}{\mathbf{w}}^{\top }\mathbf{x}+b<0, \\ 1 & \text{nếu}\phantom{\rule{4.pt}{0ex}}{\mathbf{w}}^{\top }\mathbf{x}+b\ge 0 \end{array}\right.$

---
Phương trình 5-3: Hàm mục tiêu cho bộ phân loại SVM tuyến tính biên cứng

$\def\arraystretch{1.7}\begin{array}{cc}& \underset{\mathbf{w},b}{\text{minimize}}\displaystyle \phantom{\rule{1.em}{0ex}} \frac{1}{2}{\mathbf{w}}^{\top }\mathbf{w} \\ & \text{sao cho}\phantom{\rule{1.em}{0ex}}{t}^{\left(i\right)}({\mathbf{w}}^{\top }{\mathbf{x}}^{\left(i\right)}+b)\ge 1\phantom{\rule{1.em}{0ex}}\text{với}\phantom{\rule{4.pt}{0ex}}i=1,2,\cdots ,m \end{array}$

---
Phương trình 5-4: Hàm mục tiêu cho bộ phân loại SVM tuyến tính biên mềm

$\def\arraystretch{1.7}\begin{array}{cc}& \underset{\mathbf{w},b,\zeta }{\text{minimize}}\phantom{\rule{1.em}{0ex}}{\displaystyle \frac{1}{2}}{\mathbf{w}}^{\top }\mathbf{w}+C \displaystyle \sum _{i=1}^{m}{\zeta }^{\left(i\right)} \\ & \text{sao cho}\phantom{\rule{1.em}{0ex}}{t}^{\left(i\right)}({\mathbf{w}}^{\top}{\mathbf{x}}^{\left(i\right)}+b)\ge 1-{\zeta }^{\left(i\right)}\phantom{\rule{1.em}{0ex}}\text{và}\phantom{\rule{1.em}{0ex}}{\zeta }^{\left(i\right)}\ge 0\phantom{\rule{1.em}{0ex}}\text{với}\phantom{\rule{4.pt}{0ex}}i=1,2,\cdots, m \end{array}$

---
Phương trình 5-5: Bài toán Quy hoạch Toàn phương

$\def\arraystretch{1.7}\begin{array}{ll} \underset{\mathbf{p}}{\text{Minimize}}\phantom{\rule{1.em}{0ex}}& {\displaystyle \frac{1}{2}}{\mathbf{p}}^{\top }\mathbf{H}\mathbf{p}\phantom{\rule{1.em}{0ex}}+\phantom{\rule{1.em}{0ex}}{\mathbf{f}}^{\top }\mathbf{p} \\ \text{sao cho}\phantom{\rule{1.em}{0ex}}& \mathbf{A}\mathbf{p}\le \mathbf{b} \\ \text{trong đó}\phantom{\rule{4.pt}{0ex}}& \left\{\begin{array}{ll}\mathbf{p} & \phantom{\rule{4.pt}{0ex}}\text{là vector}\phantom{\rule{4.pt}{0ex}}{n}_{p}\text{ chiều}\phantom{\rule{4.pt}{0ex}}\text{(}{n}_{p}=\text{số lượng tham số),} \\ \mathbf{H} & \phantom{\rule{4.pt}{0ex}}\text{là ma trận}\phantom{\rule{4.pt}{0ex}}{n}_{p}\times {n}_{p}\text{,} \\ \mathbf{f} & \phantom{\rule{4.pt}{0ex}}\text{là vector}\phantom{\rule{4.pt}{0ex}}{n}_{p}\text{ chiều,} \\ \mathbf{A} & \phantom{\rule{4.pt}{0ex}}\text{là ma trận}\phantom{\rule{4.pt}{0ex}}{n}_{c}\times {n}_{p}\phantom{\rule{4.pt}{0ex}}\text{}\phantom{\rule{4.pt}{0ex}}\text{(}{n}_{c}=\text{số lượng ràng buộc),} \\ \mathbf{b} & \phantom{\rule{4.pt}{0ex}}\text{là vector}\phantom{\rule{4.pt}{0ex}}{n}_{c}\text{  chiều.} \end{array}\right. \end{array}$

Lưu ý rằng biểu thức $\mathbf{A} \mathbf{p} \leq \mathbf{b}$ biểu diễn $n_{c}$ ràng buộc: $\mathbf{p}^{\top} \mathbf{a}^{(i)} \leq b^{(i)}$ với $i=1,2, \cdots, n_{c}$, trong đó $\mathbf{a}^{(i)}$ là vector chứa các phần tử nằm ở hàng thứ *i* của ma trận $\textbf{A}$ và $b^{(i)}$ là phần tử thứ *i* của $\mathbf{b}$.

Ta có thể dễ dàng chứng minh rằng nếu tham số của quy hoạch toàn phương được thiết lập như sau, ta sẽ thu được hàm mục tiêu của bộ phân loại SVM tuyến tính biên cứng:

* $n_{p}=n+1$, trong đó *n* là số lượng đặc trưng (phần $+1$ là cho hệ số điều chỉnh).
* $n_{c}=m$, trong đó $m$ là số lượng mẫu huấn luyện.
* $\mathbf{H}$ là ma trận đơn vị $n_{p} \times n_{p}$, ngoại trừ ô trên cùng bên trái có giá trị bằng 0 (để bỏ qua hệ số điều chỉnh).
* $\mathbf{f} = 0$, là vector $n_{p}$ chiều có tất cả các giá trị bằng 0.
* $\mathbf{b} = –1$, là vector $n_{c}$ chiều có tất cả các giá trị bằng –1.
* $\mathbf{a}^{(i)}=-t^{(i)} \dot{\mathbf{x}}^{(i)}$, trong đó $\dot{\mathbf{x}}^{(i)}$ là $\mathbf{x}^{(i)}$ có thêm đặc trưng ứng với hệ số điều chỉnh $\dot{\mathbf{x}}_{0}=1$.

---
Phương trình 5-6: Dạng đối ngẫu của hàm mục tiêu của SVM tuyến tính

$\def\arraystretch{1.7}\begin{array}{l} \underset{\alpha }{\text{minimize}}{\displaystyle  \phantom{\rule{1.em}{0ex}}  \frac{1}{2}} \displaystyle \sum _{i=1}^{m}\sum _{j=1}^{m}{\alpha }^{\left(i\right)}{\alpha }^{\left(j\right)}{t}^{\left(i\right)}{t}^{\left(j\right)}{{\mathbf{x}}^{\left(i\right)}}^{\top }{\mathbf{x}}^{\left(j\right)}\phantom{\rule{1.em}{0ex}}-\phantom{\rule{1.em}{0ex}}\sum _{i=1}^{m}{\alpha }^{\left(i\right)}\\ \text{sao cho}\phantom{\rule{1.em}{0ex}}{\alpha }^{\left(i\right)}\ge 0\phantom{\rule{1.em}{0ex}}\text{với}\phantom{\rule{4.pt}{0ex}}i=1,2,\cdots, m \end{array}$

Khi đã tìm được vector $\hat{\mathbf{\alpha}}$ để biểu thức này đạt cực tiểu (bằng bộ giải quy hoạch toàn phương), ta sử dụng **Phương trình 5-7** để tính $\hat{\mathbf{w}}$ và $\hat{b}$ sao cho bài toán gốc đạt cực tiểu.

Phương trình 5-7: Từ bài toán đối ngẫu về bài toán gốc

$\def\arraystretch{1.7}\begin{array}{ll}& \hat{\mathbf{w}}=\displaystyle \sum_{i=1}^{m}{\hat{\alpha }}^{\left(i\right)}{t}^{\left(i\right)}{\mathbf{x}}^{\left(i\right)} \\ & \hat{b}={\displaystyle \frac{1}{{n}_{s}}} \displaystyle\sum_{\genfrac{}{}{0pt}{}{{\scriptstyle i=1}}{{\scriptstyle {\hat{\alpha }}^{\left(i\right)}>0}}}^{m}\left({t}^{\left(i\right)}-{\hat{\mathbf{w}}}^{\top }{\mathbf{x}}^{\left(i\right)}\right) \end{array}$

---
Phương trình 5-8: Ánh xạ đa thức bậc hai

$\def\arraystretch{1.5}\phi \left(\mathbf{x}\right)=\phi \left(\left(\begin{array}{c}{x}_{1}\\ {x}_{2}\end{array}\right)\right)=\left(\begin{array}{c}{{x}_{1}}^{2}\\ \sqrt{2}{x}_{1}{x}_{2}\\ {{x}_{2}}^{2}\end{array}\right)$

---
Phương trình 5-9: Thủ thuật hạt nhân với phép ánh xạ đa thức bậc hai

$\def\arraystretch{1.5}\begin{array}{ll} \phi {\left(\mathbf{a}\right)}^{\top }\phi \left(\mathbf{b}\right)& ={\left(\begin{array}{c}{{a}_{1}}^{2}\\ \sqrt{2}\phantom{\rule{0.166667em}{0ex}}{a}_{1}{a}_{2}\\ {{a}_{2}}^{2}\end{array}\right)}^{\top }\left(\begin{array}{c}{{b}_{1}}^{2}\\ \sqrt{2}\phantom{\rule{0.166667em}{0ex}}{b}_{1}{b}_{2}\\ {{b}_{2}}^{2}\end{array}\right)={{a}_{1}}^{2}{{b}_{1}}^{2}+2{a}_{1}{b}_{1}{a}_{2}{b}_{2}+{{a}_{2}}^{2}{{b}_{2}}^{2} \\ & ={\left({a}_{1}{b}_{1}+{a}_{2}{b}_{2}\right)}^{2}={\left({\left(\begin{array}{c}{a}_{1}\\ {a}_{2}\end{array}\right)}^{\top }\left(\begin{array}{c}{b}_{1}\\ {b}_{2}\end{array}\right)\right)}^{2}={\left({\mathbf{a}}^{\top }\mathbf{b}\right)}^{2} \end{array}$

---
*Trích đoạn văn từ phần SVM Hạt nhân, trang 151 (Tập 1):*

Đây là ý tưởng quan trọng: nếu ta áp dụng phép biến đổi $\phi$ cho tất cả các mẫu huấn luyện, thì bài toán đối ngẫu (xem Phương trình 5-6) sẽ chứa tích vô hướng $\phi\left(\mathbf{x}^{(i)}\right)^{\top} \phi\left(\mathbf{x}^{(j)}\right)$. $\mathbf{X}_{\text{khôi-phục}} = \mathbf{X}_{\text{chiếu-}d} {\mathbf{W}_d}^T$

---
*Trích đoạn văn nằm ở phần LLE, trang số 200 (Tập 1):*

LLE hoạt động như sau: Với mỗi mẫu huấn luyện $\mathbf{x}^{(i)}$, thuật toán sẽ xác định $\textit{k}$ điểm gần nhất (trong đoạn mã trên $\textit{k}$ = 10), 
sau đó tái tạo lại $\mathbf{x}^{(i)}$ như một tổ hợp tuyến tính của các điểm lân cận trên. Cụ thể hơn, thuật toán sẽ tìm các trọng số $w_{i,j}$ sao cho 
bình phương khoảng cách giữa $\mathbf{x}^{(i)}$ và $ \sum_{j=1}^{m}{w}_{i,j} \mathbf{x}^{(j)}$ nhỏ nhất có thể, với giả định rằng $w_{i,j} = 0$ khi 
$\mathbf{x}^{(j)}$ không phải là một trong $\textit{k}$ điểm lân cận gần $\mathbf{x}^{(i)}$ nhất.

Vì thế, bước đầu của LLE sẽ giải một bài toán tối ưu có ràng buộc được miêu tả trong Phương trình 8-4, với $\mathbf{W}$ là ma trận trọng số chứa tất cả 
trọng số $w_{i,j}$. Điều kiện ràng buộc thứ hai chỉ đơn giản là chuẩn hóa các trọng số cho mỗi mẫu huấn luyện $\mathbf{x}^{(i)}$.


Phương trình 8-4: Bước 1 của thuật toán LLE: mô hình hóa tuyến tính các quan hệ cục bộ

$
\begin{array}{cc}& \widehat{\mathbf{W}}=\underset{\mathbf{W}}{\text{argmin}}{\displaystyle \sum _{i=1}^{m}}{\left({\mathbf{x}}^{\left(i\right)}-\displaystyle\sum _{j=1}^{m}{w}_{i,j}{\mathbf{x}}^{\left(j\right)}\right)}^{2} \\ & \text{trong}\phantom{\rule{4.pt}{0ex}}\text{đó}\phantom{\rule{4.pt}{0ex}}\left\{ \def\arraystretch{1.5} \begin{array}{ll}{w}_{i,j}=0 & \text{nếu}\phantom{\rule{4.pt}{0ex}}{\mathbf{x}}^{\left(j\right)}\phantom{\rule{4.pt}{0ex}}\text{không phải một trong}\phantom{\rule{4.pt}{0ex}}k\phantom{\rule{4.pt}{0ex}}\text{mẫu gần}\phantom{\rule{4.pt}{0ex}}{\mathbf{x}}^{\left(i\right)}\phantom{\rule{4.pt}{0ex}}\text{nhất} \\ \sum _{j=1}^{m}{w}_{i,j}=1 & \text{với}\phantom{\rule{4.pt}{0ex}}i=1,2,\cdots ,m \end{array}\right. \end{array}
$

Sau bước này, ma trận trọng số $\widehat{\mathbf{W}}$ (chứa các trọng số $\hat{w}_{i,j}$) sẽ mã hóa các quan hệ tuyến tính cục bộ giữa các mẫu huấn luyện. 
Trong bước thứ hai của LLE, các mẫu huấn luyện này sẽ được ánh xạ vào một không gian $d$-chiều (với $d<n$) trong khi cố gắng bảo toàn các quan hệ cục bộ nhiều 
nhất có thể. Nếu $\mathbf{z}^{(i)}$ là ảnh của $\mathbf{x}^{(i)}$ trong không gian $d$-chiều này, thì ta muốn bình phương khoảng cách giữa $\mathbf{z}^{(i)}$ 
và $ \sum_{j=1}^m \hat{w}_{i,j} \mathbf{z}^{(j)}$ càng nhỏ càng tốt. Ý tưởng này dẫn đến bài toán tối ưu không ràng buộc được mô tả trong Phương trình 8-5.
Bước này tương tự bước đầu tiên, nhưng thay vì giữ cho các mẫu cố định và tìm các trọng số tối ưu, ta làm ngược lại: giữ các trọng số cố định và tìm vị trí tối 
ưu cho ảnh của các mẫu trong không gian ít chiều. Lưu ý rằng $\mathbf{Z}$ là ma trận chứa tất cả các $\mathbf{z}^{(i)}$. 

Phương trình 8-5: Bước thứ hai của thuật toán LLE: giảm chiều trong khi bảo toàn các quan hệ

$
\widehat{\mathbf{Z}}=\underset{\mathbf{Z}}{\text{argmin}}{\displaystyle \sum _{i=1}^{m}}{\left({\mathbf{z}}^{\left(i\right)}-\sum _{j=1}^{m}{\widehat{w}}_{i,j}{\mathbf{z}}^{\left(j\right)}\right)}^{2}
$

## Chương 9: Các kỹ thuật Học Không giám sát

Phương trình 9-4: Độ phân kỳ KL từ $q(\mathbf{z})$ tới $p(\mathbf{z}|\mathbf{X})$

$
\def\arraystretch{2.0}
\begin{array}{ll}
D_{KL} & = \mathbb{E}_q \left[ \log \frac{\displaystyle q(\mathbf{z})}{\displaystyle p(\mathbf{z}|\mathbf{X})} \right] \\
       & = \mathbb{E}_q \left[ \log q(\mathbf{z}) - \log p(\mathbf{z}|\mathbf{X}) \right] \\
       & = \mathbb{E}_q \left[ \log q(\mathbf{z}) - \log \frac{\displaystyle p(\mathbf{z},\mathbf{X})}{\displaystyle p(\mathbf{X})} \right] \\
       & = \mathbb{E}_q \left[ \log q(\mathbf{z}) - \log p(\mathbf{z},\mathbf{X}) + \log p(\mathbf{X}) \right] \\
       & = \mathbb{E}_q \left[ \log q(\mathbf{z}) \right] - \mathbb{E}_q \left[ \log p(\mathbf{z},\mathbf{X}) \right] + \mathbb{E}_q \left[ \log p(\mathbf{X}) \right] \\
       & = \mathbb{E}_q \left[ \log p(\mathbf{X}) \right] -  \left( \mathbb{E}_q \left[ \log p(\mathbf{z},\mathbf{X}) \right] - \mathbb{E}_q \left[ \log q(\mathbf{z}) \right] \right) \\
       & = \log p(\mathbf{X}) - \text{ELBO} \\
       & \text{trong đó ELBO} = \mathbb{E}_q \left[ \log p(\mathbf{z},\mathbf{X}) \right] - \mathbb{E}_q \left[ \log q(\mathbf{z}) \right] \\
\end{array}
$

---
*Trích đoạn văn nằm trong phần AdaGrad, trang số 68 (Tập 2):*

Bước thứ hai gần giống với Hạ Gradient, với một khác biệt lớn: vector gradient được co với hệ số 
$\sqrt{\mathbf{s} + \varepsilon}$ (dấu $\oslash$ biểu diễn phép chia theo từng phần tử, và $\varepsilon$ 
là hệ số làm mượt để tránh phép chia cho 0, thường được đặt bằng $10^{-10}$). Công thức dạng vector trên 
tương đương với việc tính $\theta_i \leftarrow \theta_i - \eta \partial J (\theta) / \partial \theta_i / \sqrt{s_i + \varepsilon}$
cho mọi tham số $\theta_i$.
