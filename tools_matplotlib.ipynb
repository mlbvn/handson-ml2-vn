# Vẽ đồ thị đầu tiên của bạn

Đầu tiên chúng ta cần nạp thư viện `matplotlib`.

```python
import matplotlib
```

Matplotlib có thể xuất đồ thị bằng các thư viện đồ họa phụ trợ khác nhau, chẳng hạn như Tk, wxPython, v.v. Khi chạy python sử dụng command line, đồ thị thường được hiển thị trong một cửa sổ riêng. Trong Jupyter notebook, chúng ta có thể xuất biểu đồ trong chính notebook đó bằng cách chạy lệnh `%matplotlib inline`.

```python
%matplotlib inline
# matplotlib.use("TKAgg")  # sử dụng dòng lệnh này trong chương trình của bạn nếu bạn muốn sử dụng Tk như là một thư viện đồ họa phụ trợ
```

Bây giờ hãy vẽ đồ thị đầu tiên của chúng ta! :)

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 4, 9, 5, 3])
plt.show()
```

Đúng vậy, chỉ đơn giản là gọi hàm `plot` với một số dữ liệu, sau đó gọi hàm `show`!

Nếu hàm `plot` được cung cấp một mảng dữ liệu, thì hàm này sẽ sử dụng mảng đó làm tọa độ trên trục tung và sử dụng chỉ mục của từng điểm dữ liệu trong mảng làm tọa độ trên trục hoành.

Bạn cũng có thể cung cấp hai mảng: một cho trục hoành `x` và mảng thứ hai cho trục tung `y`:

```python
plt.plot([-3, -2, 5, 0], [1, 6, 4, 3])
plt.show()
```

Các trục tự động căn chỉnh với phạm vi của dữ liệu. Chúng ta muốn đồ thị có nhiều chỗ hơn một chút, vì vậy, hãy gọi hàm `axis` để thay đổi phạm vi của từng trục `[xmin, xmax, ymin, ymax]`.

```python
plt.plot([-3, -2, 5, 0], [1, 6, 4, 3])
plt.axis([-4, 6, 0, 7])
plt.show()
```

Bây giờ, hãy vẽ một hàm toán học. Chúng ta sử dụng hàm `linspace` của NumPy để tạo một mảng `x` chứa 500 số float từ -2 đến 2, sau đó ta tạo một mảng thứ hai `y` là bình phương của `x` (để tìm hiểu về NumPy, hãy đọc [ Hướng dẫn về NumPy](tools_numpy.ipynb)).

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show()
```

Trông nó khá đơn điệu, hãy thêm tiêu đề, nhãn x và y và vẽ một lưới.

```python
plt.plot(x, y)
plt.title("Square function")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

# Kiểu và màu Đường đồ thị

Theo mặc định, matplotlib vẽ một đường thẳng giữa các điểm liên tiếp.

```python
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0])
plt.axis([-10, 110, -10, 140])
plt.show()
```

Bạn có thêm đối số thứ 3 để thay đổi kiểu và màu của đường.
Ví dụ `"g--"` có nghĩa là "đường đứt nét màu lục".

```python
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

Bạn có thể vẽ nhiều đường trên một đồ thị một cách đơn giản: chỉ cần nhập `x1, y1, [style1], x2, y2, [style2], ...`

Ví dụ:

```python
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-", [0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

Hoặc đơn giản là gọi `plot` nhiều lần trước khi gọi `show`.

```python
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-")
plt.plot([0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

Bạn cũng có thể vẽ các điểm đơn giản thay vì các đường. Đây là một ví dụ với dấu gạch ngang xanh lục, đường chấm đỏ và khối tam giác xanh dương.
Hãy xem [tài liệu](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot) để biết danh sách đầy đủ các tùy chọn kiểu & màu.

```python
x = np.linspace(-1.4, 1.4, 30)
plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')
plt.show()
```

Hàm plot trả về một danh sách các đối tượng `Line2D` (mỗi đối tượng cho mỗi dòng). Bạn có thể đặt các thuộc tính bổ sung trên các dòng này, chẳng hạn như độ rộng của dòng, kiểu dấu gạch ngang hoặc cấp độ alpha. Xem danh sách đầy đủ các thuộc tính trong [tài liệu này](http://matplotlib.org/users/pyplot_tutorial.html#controlling-line-properties).

```python
x = np.linspace(-1.4, 1.4, 30)
line1, line2, line3 = plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')
line1.set_linewidth(3.0)
line1.set_dash_capstyle("round")
line3.set_alpha(0.2)
plt.show()
```

# Đồ thị con
Một hình matplotlib có thể chứa nhiều đồ thị con. Các đồ thị con này được sắp xếp trong một lưới. Để tạo đồ thị con, chỉ cần gọi hàm `subplot`, chỉ định số lượng hàng và cột trong hình, và chỉ mục của đồ thị con mà bạn muốn vẽ (bắt đầu từ 1, sau đó từ trái sang phải và từ trên xuống dưới ). Lưu ý rằng pyplot theo dõi đồ thị con hiện đang hoạt động (mà bạn có thể lấy tham chiếu bằng cách gọi `plt.gca()`), vì vậy khi bạn gọi hàm `plot`, nó sẽ dựa trên đồ thị con trong tình trạng *đang hoạt động*.


```python
x = np.linspace(-1.4, 1.4, 30)
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot = top left
plt.plot(x, x)
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot = top right
plt.plot(x, x**2)
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, 3rd subplot = bottow left
plt.plot(x, x**3)
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, 4th subplot = bottom right
plt.plot(x, x**4)
plt.show()
```

* Lưu ý rằng `subplot(223)` là viết tắt của `subplot(2, 2, 3)`.

Thật dễ dàng để tạo các đồ thị con trải dài trên nhiều ô lưới như sau:

```python
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot = top left
plt.plot(x, x)
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot = top right
plt.plot(x, x**2)
plt.subplot(2, 1, 2)  # 2 rows, *1* column, 2nd subplot = bottom
plt.plot(x, x**3)
plt.show()
```

Nếu bạn cần định vị đồ thị con phức tạp hơn, bạn có thể sử dụng `subplot2grid` thay vì `subplot`. Bạn chỉ định số lượng hàng và cột trong lưới, sau đó là vị trí ô con của bạn trong lưới đó (trên cùng bên trái = (0,0)) và tùy chọn bao nhiêu hàng và/hoặc cột kéo dài. Ví dụ:

```python
plt.subplot2grid((3,3), (0, 0), rowspan=2, colspan=2)
plt.plot(x, x**2)
plt.subplot2grid((3,3), (0, 2))
plt.plot(x, x**3)
plt.subplot2grid((3,3), (1, 2), rowspan=2)
plt.plot(x, x**4)
plt.subplot2grid((3,3), (2, 0), colspan=2)
plt.plot(x, x**5)
plt.show()
```

Nếu bạn cần linh hoạt hơn nữa trong việc định vị đồ thị con, tham khảo [tài liệu GridSpec](http://matplotlib.org/users/gridspec.html).

# Đa Đồ thị
Ta cũng có thể vẽ nhiều đồ thị. Mỗi đồ thị có thể chứa một hoặc nhiều đồ thị con. Theo mặc định, matplotlib tự động tạo `figure(1)`. Khi bạn chuyển đổi đồ thị, pyplot theo dõi hình hiện đang hoạt động (bạn có thể lấy tham chiếu bằng cách gọi `plt.gcf()`) và đồ thị con đang hoạt động của đồ thị đó trở thành đồ thị con hiện tại.

```python
x = np.linspace(-1.4, 1.4, 30)

plt.figure(1)
plt.subplot(211)
plt.plot(x, x**2)
plt.title("Square and Cube")
plt.subplot(212)
plt.plot(x, x**3)

plt.figure(2, figsize=(10, 5))
plt.subplot(121)
plt.plot(x, x**4)
plt.title("y = x**4")
plt.subplot(122)
plt.plot(x, x**5)
plt.title("y = x**5")

plt.figure(1)      # back to figure 1, current subplot is 212 (bottom)
plt.plot(x, -x**3, "r:")

plt.show()
```

# Máy trạng thái của Pyplot: ẩn *và* rõ ràng
Cho đến nay, chúng ta đã sử dụng máy trạng thái của Pyplot để theo dõi đồ thị con hiện đang hoạt động. Mỗi khi bạn gọi hàm `plot`, pyplot chỉ vẽ trên subplot hiện đang hoạt động. Nó cũng thực hiện một số "phép thuật" khác, chẳng hạn như tự động tạo một hình và một đồ thị con khi bạn gọi `plot`, nếu chúng chưa tồn tại. Phép thuật này thuận tiện trong môi trường tương tác (chẳng hạn như Jupyter).

Nhưng khi bạn đang viết một chương trình, *sự rõ ràng sẽ tốt hơn những phép ẩn này*. Mã lập trình rõ ràng thường dễ gỡ lỗi và bảo trì hơn, và nếu bạn không tin tôi chỉ cần đọc quy tắc thứ 2 trong Zen of Python:

```python
import this
```

May mắn thay, Pyplot cho phép bạn bỏ qua hoàn toàn máy trạng thái, vì vậy bạn có thể viết mã rõ ràng đẹp mắt. Chỉ cần gọi hàm `subplots` và sử dụng đối tượng hình và danh sách các đối tượng trục được trả về. Không còn phép thuật nào nữa! Ví dụ:

```python
x = np.linspace(-2, 2, 200)
fig1, (ax_top, ax_bottom) = plt.subplots(2, 1, sharex=True)
fig1.set_size_inches(10,5)
line1, line2 = ax_top.plot(x, np.sin(3*x**2), "r-", x, np.cos(5*x**2), "b-")
line3, = ax_bottom.plot(x, np.sin(3*x), "r-")
ax_top.grid(True)

fig2, ax = plt.subplots(1, 1)
ax.plot(x, x**2)
plt.show()
```

Để nhất quán, chúng ta sẽ tiếp tục sử dụng máy trạng thái của pyplot trong phần còn lại của hướng dẫn này, nhưng chúng tôi khuyên bạn nên sử dụng giao diện hướng đối tượng trong các chương trình của mình.

# Pylab *vs* Pyplot *vs* Matplotlib

Có một số nhầm lẫn xung quanh mối quan hệ giữa pylab, pyplot và matplotlib. Thật đơn giản: matplotlib là thư viện đầy đủ, nó chứa mọi thứ kể cả pylab và pyplot.

Pyplot cung cấp một số công cụ để vẽ biểu đồ, bao gồm giao diện máy trạng thái cho thư viện vẽ đồ thị hướng đối tượng bên dưới.

Pylab là một mô-đun tiện lợi nhập matplotlib.pyplot và NumPy trong một không gian tên duy nhất. Bạn sẽ tìm thấy nhiều ví dụ sử dụng pylab, nhưng nó không còn được khuyến nghị nữa (vì cách nhập *rõ ràng* tốt hơn nhập *ẩn*).

# Vẽ văn bản
Bạn có thể gọi `text` để thêm văn bản vào bất kỳ vị trí nào trong biểu đồ. Chỉ cần chỉ định tọa độ ngang, dọc, nội dung văn bản, và tùy chọn một số thuộc tính bổ sung. Mọi văn bản trong matplotlib có thể chứa biểu thức phương trình TeX, xem [tài liệu](http://matplotlib.org/users/mathtext.html) để biết thêm chi tiết.

```python
x = np.linspace(-1.5, 1.5, 30)
px = 0.8
py = px**2

plt.plot(x, x**2, "b-", px, py, "ro")

plt.text(0, 1.5, "Square function\n$y = x^2$", fontsize=20, color='blue', horizontalalignment="center")
plt.text(px - 0.08, py, "Beautiful point", ha="right", weight="heavy")
plt.text(px, py, "x = %0.2f\ny = %0.2f"%(px, py), rotation=50, color='gray')

plt.show()
```

* Lưu ý: `ha` là bí danh của `horizontalalignment`

Để biết thêm các thuộc tính văn bản, hãy truy cập [tài liệu](http://matplotlib.org/users/text_props.html#text-properties).

Việc chú thích các thành phần của biểu đồ  diễn ra khá thường xuyên, chẳng hạn như Beautiful point ở trên. Chức năng `chú thích` làm cho điều này trở nên dễ dàng: chỉ cần cho biết vị trí của điểm ưa thích và vị trí của văn bản, cùng với một số thuộc tính bổ sung tùy chọn cho văn bản và mũi tên.

```python
plt.plot(x, x**2, px, py, "ro")
plt.annotate("Beautiful point", xy=(px, py), xytext=(px-1.3,py+0.5),
                           color="green", weight="heavy", fontsize=14,
                           arrowprops={"facecolor": "lightgreen"})
plt.show()
```

Bạn cũng có thể thêm khung chứa xung quanh văn bản của mình bằng cách sử dụng thuộc tính `bbox`:

```python
plt.plot(x, x**2, px, py, "ro")

bbox_props = dict(boxstyle="rarrow,pad=0.3", ec="b", lw=2, fc="lightblue")
plt.text(px-0.2, py, "Beautiful point", bbox=bbox_props, ha="right")

bbox_props = dict(boxstyle="round4,pad=1,rounding_size=0.2", ec="black", fc="#EEEEFF", lw=5)
plt.text(0, 1.5, "Square function\n$y = x^2$", fontsize=20, color='black', ha="center", bbox=bbox_props)

plt.show()
```



Để tăng phần thú vị, nếu bạn muốn có một biểu đồ kiểu [xkcd](http://xkcd.com), chỉ cần vẽ trong phần `with plt.xkcd()`:

# Chú thích

# Biểu diễn Phi tuyến
Matplotlib hỗ trợ các thang đo phi tuyến tính, chẳng hạn như thang đo logarit hoặc logit.

```python
x = np.linspace(0.1, 15, 500)
y = x**3/np.exp(2*x)

plt.figure(1)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

plt.figure(2)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

plt.figure(3)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)

plt.figure(4)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.05)
plt.title('symlog')
plt.grid(True)

plt.show()
```

# Tick và ticker
Các trục có những dấu nhỏ gọi là "tick". Nói chính xác, "tick" là *vị trí* của các dấu (ví dụ: (-1, 0, 1)), "đường tick" là các đường nhỏ được vẽ tại các vị trí đó, "nhãn tick" là các nhãn được vẽ tiếp theo đến các dòng tick và "ticker" là đối tượng có khả năng quyết định vị trí đặt tick. Các ticker mặc định thường thực hiện khá tốt việc đặt ~5 đến 8 mã đánh dấu ở khoảng cách hợp lý với nhau.

Nhưng đôi khi bạn cần nhiều kiểm soát hơn (ví dụ: có quá nhiều nhãn đánh dấu trên biểu đồ logit ở trên). May mắn thay, matplotlib cung cấp cho bạn toàn quyền kiểm soát các tick. Bạn thậm chí có thể kích hoạt các tick nhỏ.


```python
x = np.linspace(-2, 2, 100)

plt.figure(1, figsize=(15,10))
plt.subplot(131)
plt.plot(x, x**3)
plt.grid(True)
plt.title("Default ticks")

ax = plt.subplot(132)
plt.plot(x, x**3)
ax.xaxis.set_ticks(np.arange(-2, 2, 1))
plt.grid(True)
plt.title("Manual ticks on the x-axis")

ax = plt.subplot(133)
plt.plot(x, x**3)
plt.minorticks_on()
ax.tick_params(axis='x', which='minor', bottom=False)
ax.xaxis.set_ticks([-2, 0, 1, 2])
ax.yaxis.set_ticks(np.arange(-5, 5, 1))
ax.yaxis.set_ticklabels(["min", -4, -3, -2, -1, 0, 1, 2, 3, "max"])
plt.title("Manual ticks and tick labels\n(plus minor ticks) on the y-axis")


plt.grid(True)

plt.show()
```

# Phép chiếu Cực
Ta có thể dễ dàng vẽ biểu đồ cực bằng cách thiết lập thuộc tính `projection` thành `"polar"` khi tạo đồ thị con.

```python
radius = 1
theta = np.linspace(0, 2*np.pi*radius, 1000)

plt.subplot(111, projection='polar')
plt.plot(theta, np.sin(5*theta), "g-")
plt.plot(theta, 0.5*np.cos(20*theta), "b-")
plt.show()
```

# Phép chiếu 3D

Vẽ đồ thị 3D khá đơn giản. Ta chỉ cần nạp `Axes3D` để đăng ký phép chiếu `"3d"`. Sau đó, tạo một đồ thị con và thiết lập `projection` thành `"3d"`. Việc này trả về một đối tượng `Axes3DSubplot` mà bạn có thể sử dụng để gọi `plot_surface`, cung cấp các tọa độ x, y và z cùng với các thuộc tính tùy chọn.

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()
```

Một cách khác để hiển thị cùng dữ liệu này là *thông qua* biểu đồ đường viền.


```python
plt.contourf(X, Y, Z, cmap=matplotlib.cm.coolwarm)
plt.colorbar()
plt.show()
```

# Biểu đồ phân tán

Để vẽ biểu đồ phân tán, ta chỉ cần cung cấp tọa độ x và y của các điểm.

```python
from numpy.random import rand
x, y = rand(2, 100)
plt.scatter(x, y)
plt.show()
```

Bạn cũng có thể tùy chọn thiết lập kích cỡ của từng điểm.

```python
x, y, scale = rand(3, 100)
scale = 500 * scale ** 5
plt.scatter(x, y, s=scale)
plt.show()
```

Và như thường lệ, có một số thuộc tính khác mà bạn cũng có thể thiết lập, chẳng hạn như màu tô và độ alpha.

```python
for color in ['red', 'green', 'blue']:
    n = 100
    x, y = rand(2, n)
    scale = 500.0 * rand(n) ** 5
    plt.scatter(x, y, s=scale, c=color, alpha=0.3, edgecolors='blue')

plt.grid(True)

plt.show()
```

# Đường
Bạn có thể vẽ các đường đơn giản bằng cách sử dụng chức năng `plot`, như chúng ta đã làm cho đến nay. Tuy nhiên, sẽ thuận tiện hơn khi tạo một hàm tiện ích vẽ một đường (dường như) vô hạn trên biểu đồ, biết trước hệ số góc và một giao điểm. Bạn cũng có thể sử dụng các hàm `hlines` và `vlines` để vẽ các đoạn đường ngang và dọc.
Ví dụ:

```python
from numpy.random import randn

def plot_line(axis, slope, intercept, **kargs):
    xmin, xmax = axis.get_xlim()
    plt.plot([xmin, xmax], [xmin*slope+intercept, xmax*slope+intercept], **kargs)

x = randn(1000)
y = 0.5*x + 5 + randn(1000)*2
plt.axis([-2.5, 2.5, -5, 15])
plt.scatter(x, y, alpha=0.2)
plt.plot(1, 0, "ro")
plt.vlines(1, -5, 0, color="red")
plt.hlines(0, -2.5, 1, color="red")
plot_line(axis=plt.gca(), slope=0.5, intercept=5, color="magenta")
plt.grid(True)
plt.show()
```

# Hình ảnh
Đọc, tạo và vẽ đồ thị hình ảnh trong matplotlib khá đơn giản.

Để đọc một hình ảnh, chỉ cần nhập mô-đun `matplotlib.image` và gọi hàm `imread` của nó, chuyển cho nó tên tệp (hoặc đối tượng tệp). Thao tác này trả về dữ liệu hình ảnh dưới dạng một mảng NumPy. Hãy thử điều này với hình ảnh `my_square_function.png` mà chúng tôi đã lưu trước đó.

```python
import matplotlib.image as mpimg

img = mpimg.imread('my_square_function.png')
print(img.shape, img.dtype)
```

Chúng ta đã nạp một hình ảnh 288x432. Mỗi điểm ảnh được biểu thị bằng một mảng 4 phần tử: đỏ, xanh lá cây, xanh dương và các cấp độ alpha, được lưu trữ dưới dạng số thực 32 bit trong khoảng từ 0 đến 1. Bây giờ, tất cả những gì chúng ta cần làm là gọi `imshow`:

```python
plt.imshow(img)
plt.show()
```

Tadaaa! Bạn có thể muốn ẩn các trục khi hiển thị hình ảnh:

```python
plt.imshow(img)
plt.axis('off')
plt.show()
```

Tạo ảnh của riêng bạn cũng rất dễ dàng:

```python
img = np.arange(100*100).reshape(100, 100)
print(img)
plt.imshow(img)
plt.show()
```

Vì chúng ta không cung cấp các mức RGB nên hàm `imshow` sẽ tự động ánh xạ các giá trị thành một dải màu. Theo mặc định, dải màu chuyển từ màu lam (đối với giá trị thấp) sang màu đỏ (đối với giá trị cao), nhưng bạn có thể chọn bản đồ màu khác. Ví dụ:

```python
plt.imshow(img, cmap="hot")
plt.show()
```

Bạn cũng có thể tạo một hình ảnh RGB trực tiếp:

```python
img = np.empty((20,30,3))
img[:, :10] = [0, 0, 0.6]
img[:, 10:20] = [1, 1, 1]
img[:, 20:] = [0.6, 0, 0]
plt.imshow(img)
plt.show()
```

Vì mảng `img` khá nhỏ (20x30), nên khi hàm `imshow` hiển thị mảng đó, nó sẽ phóng to hình ảnh theo kích thước của đồ thị. Hãy tưởng tượng kéo dài hình ảnh gốc từ đó để lại khoảng trống giữa các điểm ảnh gốc. Làm thế nào để `imshow` điền vào chỗ trống? Chà, theo mặc định, nó chỉ tô màu từng pixel trống bằng cách sử dụng màu của pixel không trống gần nhất. Kỹ thuật này có thể dẫn đến các ảnh mờ. Nếu muốn, bạn có thể sử dụng một phương pháp nội suy khác, chẳng hạn như [nội suy song tuyến tính](https://vi.wikipedia.org/wiki/Bilinear_interpolation) để điền vào các pixel trống. Điều này dẫn đến các cạnh bị làm mờ, tuy nhiên trong một số trường hợp sẽ trông đẹp hơn:

```python
plt.imshow(img, interpolation="bilinear")
plt.show()
```

# Ảnh động
Mặc dù matplotlib chủ yếu được sử dụng để tạo hình ảnh nhưng nó cũng có khả năng hiển thị ảnh động. Trước tiên, bạn cần nạp `matplotlib.animation`.

```python
import matplotlib.animation as animation
```

Trong ví dụ sau, chúng ta bắt đầu bằng cách tạo các điểm dữ liệu, sau đó tạo một đồ thị trống. Ta định nghĩa chức năng cập nhật được gọi ở mỗi lần lặp lại ảnh động, và cuối cùng thêm ảnh động vào đồ thị bằng cách tạo thực thể `FuncAnimation`.

Hàm khởi tạo `FuncAnimation` nhận một đồ thị, một hàm cập nhật và các đối số tùy chọn. Chúng ta muốn tạo một ảnh động dài 50 khung hình, với 100 mili giây giữa mỗi khung hình. Tại mỗi vòng lặp, `FuncAnimation` gọi hàm cập nhật và chuyển cho nó số khung `num` (từ 0 đến 49 trong trường hợp này) theo sau là các đối số bổ sung mà chúng ta đã chỉ định bằng `fargs`.

Chức năng cập nhật chỉ đơn giản là đặt dữ liệu đường thành các điểm dữ liệu `num` đầu tiên (để dữ liệu được vẽ dần dần) và để tăng tính giải trí, ta cũng có thể thêm một số  giá trị ngẫu nhiên nhỏ vào mỗi điểm dữ liệu để  các đường trông có vẻ lắc lư.

```python
x = np.linspace(-1, 1, 100)
y = np.sin(x**2*25)
data = np.array([x, y])

fig = plt.figure()
line, = plt.plot([], [], "r-") # start with an empty plot
plt.axis([-1.1, 1.1, -1.1, 1.1])
plt.plot([-0.5, 0.5], [0, 0], "b-", [0, 0], [-0.5, 0.5], "b-", 0, 0, "ro")
plt.grid(True)
plt.title("Marvelous animation")

# this function will be called at every iteration
def update_line(num, data, line):
    line.set_data(data[..., :num] + np.random.rand(2, num) / 25)  # we only plot the first `num` data points.
    return line,

line_ani = animation.FuncAnimation(fig, update_line, frames=50, fargs=(data, line), interval=100)
plt.close() # call close() to avoid displaying the static plot
```

Tiếp theo, hãy hiển thị hình ảnh động. Một tùy chọn là chuyển đổi nó thành mã HTML5 (sử dụng thẻ `<video>`) và hiển thị mã này bằng cách sử dụng `IPython.display.HTML`:

```python
from IPython.display import HTML

HTML(line_ani.to_html5_video())
```

Ngoài ra, chúng ta có thể hiển thị ảnh động bằng tiện ích tương tác HTML/Javascript nhỏ xinh:

```python
HTML(line_ani.to_jshtml())
```

Bạn có thể  cấu hình Matplotlib để sử dụng tiện ích này theo mặc định khi hiển thị ảnh động:

```python
matplotlib.rc('animation', html='jshtml')
```

Sau đó, bạn không cần thiết cần sử dụng `IPython.display.HTML` nữa:

```python
animation.FuncAnimation(fig, update_line, frames=50, fargs=(data, line), interval=100)
```

# Lưu hình động vào tệp video
Matplotlib dựa vào các thư viện của bên thứ 3 để tạo các video như [FFMPEG](https://www.ffmpeg.org/) hoặc [ImageMagick](https://imagemagick.org/). Trong ví dụ này, chúng ta sẽ sử dụng FFMPEG, vì vậy hãy đảm bảo ta cài đặt nó trước. Để lưu hoạt ảnh sang định dạng GIF, bạn sẽ cần ImageMagick.

```python
Writer = animation.writers['ffmpeg']
writer = Writer(fps=15, metadata=dict(artist='Me'), bitrate=1800)
line_ani.save('my_wiggly_animation.mp4', writer=writer)
```
